```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python

class Net(nn.Module):
    def __init__(self, in_c, out_c):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(in_c, out_c, kernel_size=3)
        self.conv2 = nn.Conv2d(out_c, out_c, kernel_size=3)
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        return x
    

model = Net(3, 6)
print(model)

```
